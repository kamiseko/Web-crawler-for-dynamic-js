```python
import json
import requests
import pandas as pd
import urllib2
```

```python
# 爬取数据
url = 'https://www.jisilu.cn/data/cbnew/cb_list/'
data = urllib2.urlopen(url)
new = data.read()
newdata = json.loads(new)[u'rows']
```

```python
newdata.__len__()
```

```python
# json格式转dataframe
columnsName = [u'转债代码',u'正股代码',u'转股价',u'到期税后收益',u'评级',u'发行时间',u'上市时间',u'转股起始日',u'到期时间',
               u'到期年限',u'回售价格',u'强赎价格']
dataList = []
for item in newdata:
    newitem = item[u'cell']
    itemList = [newitem['bond_id'],newitem['stock_id'],newitem['convert_price'],newitem['ytm_rt_tax'],
                newitem['rating_cd'],newitem['issue_dt'],newitem['list_dt'],newitem['convert_dt'],
                newitem['maturity_dt'],newitem['year_left'],newitem['put_convert_price'],newitem['force_redeem_price']]
    dataList.append(itemList)
```

```python
cbinfo = pd.DataFrame(dataList,columns =columnsName)
cbinfo[u'正股代码完整'] = cbinfo[u'正股代码'].map(lambda x: x[2:] +'.'+x[:2].upper())
cbinfo[u'转债代码完整'] = cbinfo.apply(lambda x: x[u'转债代码'] + x[u'正股代码完整'][-3:],axis=1)
cbinfo
```
